### Blending Tank Simulation
Simple process modeled by a pair of ODEs
\begin{align*}
\frac{dV}{dt} & = \frac{1}{\rho}(w_1 + w_2 - w)\\
\frac{dx}{dt} & = \frac{1}{\rho V}(w_1 (x_1 - x) + w_2 (x_2 - x))
\end{align*}

Initializing graphics display and importing needed libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
from scipy.integrate import odeint
```

Establish parameter values

```python
rho = 900.0

w1 = 500.0 # Stream 1, kg/min
w2 = 200.0 # Stream 2, kg/min
w = 650.0 # Set outflow

x1 = 0.4 # Mass frac stream 1
x2 = 0.75 # Mass frac stream 2
```

Function to compute RHS of the differential Eqs

```python
def RHS(y,t):
    V,x = y
    dVdt= (w1+w2-w)/rho
    dxdt= (w1*(x1-x)+w2*(x2-x))/(rho*V)
    return [dVdt,dxdt]
```

Initial conditions, time grid and integration

```python
V0 = 2.0 # Initial volume
x0 = 0.0 # Initial composition, mass frac

t=np.linspace(0,10.0)
y=odeint(RHS,[V0,x0],t)
```

Visualization

```python
plt.plot(t,y)
plt.xlabel('Time [min]')
plt.ylabel('Volume, Composition')
plt.legend(['Volume','Composition'])
plt.ylim(0,3)
plt.grid()
```

Steady State Analysis

```python
w1 = 500.0   # stream 1, kg/min
w2 = 200.0   # stream 2, kg/min
w = w1+w2 # Outflow = Inflow, steady state assumption

x1 = 0.4     # composition stream 1, mass fraction
x2 = 0.75    # composition stream 2, mass fraction

x = (w1*x1 + w2*x2)/(w)

print('Steady State Composition =', x, "[mass fraction]")
```
