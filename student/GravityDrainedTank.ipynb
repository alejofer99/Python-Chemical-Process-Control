### Gravity Drained Tank
Nonlinear process model for the liquid level in a gravity drained tank

Initializing graphics display and importing needed libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
from scipy.integrate import odeint
```

Defining parameters

```python
Cv = 0.1 # Outlet valve constant [cubic meters/min/meter^1/2]
A = 1.0 # Tank area [meter^2] 
```

Functions for the RHS of the Differential equations: <br />
$A\frac{dh}{dt} = q_{in}(t) - q_{out}(t)$ <br />
$q_{out}(h) = C_v\sqrt{h}$ <br />
$\frac{dh}{dt} = - \frac{C_V}{A}\sqrt{h} + \frac{1}{A}q_{in}(t)$

```python
def qin(t):
    return 0.15

def deriv(h,t):
    return qin(t)/A - Cv*np.sqrt(h)/A
```

Initial condition, time grid and integration of the differential eq.

```python
IC = [0.0]
t = np.linspace(0,200,101)
h = odeint(deriv,IC,t)
```

Visualize and Analyze results

```python
plt.plot(t,h)

plt.xlabel("minutes")
plt.ylabel("meters")
plt.title("Gravity-Drained Tank")
plt.legend(["Liquid Level"]);
```
